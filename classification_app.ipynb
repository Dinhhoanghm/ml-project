```python
import tkinter as tk
from tkinter import filedialog, Label, Button, OptionMenu
from PIL import Image, ImageTk
import tensorflow as tf
import numpy as np
from keras.models import load_model

# Hàm xử lý ảnh
def preprocess_image(image, model_type):
    if model_type == 'CNN':
        IMAGE_SIZE = 200
    elif model_type == 'ANN':
        IMAGE_SIZE = 112
    elif model_type == 'VGG':
        IMAGE_SIZE = 224
    elif model_type == 'Xception':
        IMAGE_SIZE = 224
    elif model_type == 'Inception':
        IMAGE_SIZE = 224
    
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [IMAGE_SIZE, IMAGE_SIZE])
    image /= 255.0  # normalize to [0,1] range
    return image

# Hàm tải và xử lý ảnh
def load_and_preprocess_image(path, model_type):
    image = tf.io.read_file(path)
    return preprocess_image(image, model_type)

# Hàm dự đoán ảnh
def predict_image(model, file_path, model_type):
    image = load_and_preprocess_image(file_path, model_type)
    image = np.expand_dims(image, axis=0)
    pred = model.predict(image)
    if pred[0] > 0.5:
        label = 'Dog nè'
    else:
        label = 'Cat nè'
    return label
def load_custom_model(model_path):
    return load_model(model_path)
# Hàm upload ảnh
def upload_image():
    try:
        file_path = filedialog.askopenfilename()
        uploaded = Image.open(file_path)
        uploaded.thumbnail((400, 400))
        im = ImageTk.PhotoImage(uploaded)
        sign_image.configure(image=im)
        sign_image.image = im
        label.configure(text='')
        show_classify_button(model, file_path, model_var.get())
    except Exception as e:
        print(e)

# Hàm phân loại ảnh
def classify(model, file_path, model_type):
    label_text = predict_image(model, file_path, model_type)
    label.configure(text=label_text, foreground='#011638')

# Hàm hiển thị nút phân loại
def show_classify_button(model, file_path, model_type):
    classify_b = Button(top, text="Classify Image", command=lambda: classify(model, file_path, model_type), padx=10, pady=5)
    classify_b.configure(background='#364156', foreground='white', font=('arial', 10, 'bold'))
    classify_b.place(relx=0.79, rely=0.46)

# Hàm chuyển đổi mô hình
def switch_model(model_name):
    global model
    model_path = model_paths[model_name]
    model = load_custom_model(model_path)

# Khởi tạo GUI
top = tk.Tk()
top.geometry('800x600')
top.title('CatsVSDogs Classification')
top.configure(background='#CDCDCD')

label = Label(top, background='#CDCDCD', font=('arial', 15, 'bold'))
sign_image = Label(top)

upload = Button(top, text="Upload an image", command=upload_image, padx=10, pady=5)
upload.configure(background='#364156', foreground='white', font=('arial', 10, 'bold'))
upload.pack(side=tk.BOTTOM, pady=50)
sign_image.pack(side=tk.BOTTOM, expand=True)
label.pack(side=tk.BOTTOM, expand=True)
heading = Label(top, text="CatsVSDogs Classification", pady=20, font=('arial', 20, 'bold'))
heading.configure(background='#CDCDCD', foreground='#364156')
heading.pack()

# Danh sách các mô hình và đường dẫn
model_paths = {
    'CNN': r'/code/jupyter/best_cnn_model_50epoch.h5',
    'ANN': r'/code/jupyter/Downloads\ann_model.h5',
    'VGG':r'/code/jupyter/Downloads\vgg_model.h5',
    'Inception':r'/code/jupyter/inception_model.h5',
    'Xception':r'/code/jupyter/xception_model.h5'
    # Thêm các mô hình khác nếu cần
}

# Dropdown menu để chọn mô hình
model_var = tk.StringVar(top)
model_var.set('CNN')  # Mô hình mặc định là CNN

model_menu = OptionMenu(top, model_var, *model_paths.keys(), command=switch_model)
model_menu.configure(background='#364156', foreground='white', font=('arial', 10, 'bold'))
model_menu.pack(side=tk.TOP, pady=10)

# Tải mô hình mặc định (CNN)
model = load_custom_model(model_paths[model_var.get()])

top.mainloop()
```
